```python
import pandas as pd

# Carga tus datos en un DataFrame de Pandas
# Reemplaza 'ruta_del_archivo.csv' con la ruta de tu archivo de datos
datos = pd.read_csv(r'C:\Users\kervi\OneDrive\Escritorio\DATA\proyecto modulo 2\17 septiembre 2023\listings17.csv')

def porcentaje_nulos(datos):
    # Calcula el total de valores nulos por columna
    nulos_por_columna = datos.isnull().sum()
    
    # Calcula el total de filas
    total_filas = len(datos)
    
    # Calcula el porcentaje de nulos por columna
    porcentaje_nulos_por_columna = (nulos_por_columna / total_filas) * 100
    
    return porcentaje_nulos_por_columna

# Calcula el porcentaje de nulos por columna en tus datos
porcentaje_nulos_por_columna = porcentaje_nulos(datos)

# Imprime los resultados
print("Porcentaje de valores nulos por columna en tus datos:")
print(porcentaje_nulos_por_columna)
```

```python
# Nombre de la columna específica
columna_especifica = 'last_searched'

# Número de valores nulos en la columna específica
num_nulos = datos[columna_especifica].isnull().sum()
print("Número de valores nulos en la columna '{}': {}".format(columna_especifica, num_nulos))

# Número total de filas en la columna
total_filas = len(datos)

# Porcentaje de filas con atributos nulos en la columna
porcentaje_nulos = (num_nulos / total_filas) * 100
print("Porcentaje de filas con atributos nulos en la columna '{}': {:.2f}%".format(columna_especifica, porcentaje_nulos))
```

```python
import pandas as pd

# Carga tus datos en un DataFrame de Pandas
# Reemplaza 'ruta_del_archivo.csv' con la ruta de tu archivo de datos
datos = pd.read_csv(r'C:\Users\kervi\OneDrive\Escritorio\DATA\proyecto modulo 2\17 septiembre 2023\listings17.csv')

def nulos_y_porcentaje(datos):
    # Calcula el número de valores nulos por columna
    nulos_por_columna = datos.isnull().sum()
    
    # Calcula el total de filas
    total_filas = len(datos)
    
    # Calcula el porcentaje de nulos por columna
    porcentaje_nulos_por_columna = (nulos_por_columna / total_filas) * 100
    
    # Calcula el porcentaje de filas con valores nulos en su respectiva columna
    porcentaje_filas_con_nulos_por_columna = (nulos_por_columna / total_filas) * 100
    
    return nulos_por_columna, porcentaje_filas_con_nulos_por_columna

# Calcula el número de valores nulos por columna y el porcentaje de filas con valores nulos por columna
nulos_por_columna, porcentaje_filas_con_nulos_por_columna = nulos_y_porcentaje(datos)

# Imprime los resultados
print("Número de valores nulos por columna:")
print(nulos_por_columna)
print("\nPorcentaje de filas con valores nulos en su respectiva columna:")
print(porcentaje_filas_con_nulos_por_columna)
```

```python
# Define la función para eliminar columnas con valores nulos mayores al 60%
def eliminar_columnas_con_nulos(df, umbral=60):
    # Calcula el porcentaje de valores nulos por columna
    porcentaje_nulos_por_columna = (df.isnull().sum() / len(df)) * 100
    
    # Encuentra las columnas que tienen valores nulos mayores al umbral especificado
    columnas_a_eliminar = porcentaje_nulos_por_columna[porcentaje_nulos_por_columna > umbral].index
    
    # Elimina las columnas con valores nulos mayores al umbral
    df_sin_columnas_nulas = df.drop(columns=columnas_a_eliminar)
    
    return df_sin_columnas_nulas

# Elimina las columnas con valores nulos mayores al 60%
datos_sin_columnas_nulas = eliminar_columnas_con_nulos(datos, umbral=60)

# Imprime los datos restantes para visualización
print("Datos después de eliminar columnas con valores nulos mayores al 60%:")
print(datos_sin_columnas_nulas)
```
